```python
import pandas as pd
import numpy as np
```

```python
input_df = pd.read_csv('Inputdata - Sheet1.csv')

input_df.info()
input_df['Result Type'] = input_df['Result Type'].fillna('Unknown')
input_df.fillna(0, inplace = True)

input_df
```

```python
output_df = pd.read_csv('OutputFormat - Sheet1.csv')
output_df = output_df.iloc[:,:10]

output_df
```

```python
def Campaign_Results(input_df):
    
    date_df = pd.DataFrame(input_df['Day'])
    date_df.rename(columns = {'Day' : 'Date'},inplace = True)
    date_df['Date'] = pd.to_datetime(date_df['Date'])
    date_df['Date'] = date_df['Date'].apply(lambda x: x.strftime('%B %d (%A)'))
    date_df = date_df.sort_values(by = 'Date').reset_index(drop = True)
    date_df
    
    columns = ['Date', 'Campaign Name', 'AD Spent', 'No.of Leads', 'Cost Per Lead',
       'CTR', 'Link Click', 'LP Views', 'member passed', 'LP Conversion %']

    output_df = pd.DataFrame(columns = columns)

    output_df['Date'] = date_df
    output_df['Campaign Name'] = input_df['Campaign name']
    output_df['AD Spent'] = input_df['Amount spent (INR)']
    output_df['No.of Leads'] = input_df['Results']
    output_df['Cost Per Lead'] = (output_df['AD Spent'] / output_df['No.of Leads']).replace(np.inf,0)
    output_df['CTR'] = input_df['CTR (link click-through rate)'].apply(lambda x: '{:.2}%'.format(x))
    output_df['Link Click'] = input_df['Link clicks']
    output_df['LP Views'] = input_df['Landing page views']
    output_df['member passed'] = (output_df['LP Views'] / output_df['Link Click']) * 100
    output_df['LP Conversion %'] = (output_df['No.of Leads'] / output_df['LP Views']).apply(
        lambda x: '{:.2%}'.format(x)).replace(to_replace = '.*n.*', value = 0,regex = True)
    output_df.fillna('Unknown',inplace = True)
    
    return output_df

output_df = Campaign_Results(input_df)
output_df
```
